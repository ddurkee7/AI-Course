Functions mapping R^2 to R:

> Local min: -1

z = cos(x) * (3x / (1 + 4y))
> Local max: 19

Setting up:

```
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
```

Create functions 1, 2, and 3:

```
def f1(x, y):
  return x + y

def f2(x, y):
  return np.cos(np.sqrt(x) + y**2 - 2 * y)

def f3(x, y):
  return np.cos(x) * ((3 * x) / (1 + (4 * y)))
```

Create values of 0 to 10 for x and y, then run them through function 1, plot, and find maximum:

```
x = np.linspace(0, 10)
y = np.linspace(0, 10)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

plt.contourf(X, Y, Z, 20, cmap='OrRd')
plt.colorbar();
```

3D plot:

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

2D and 3D (at different viewing angle) plots for f2:

```
Z = f2(X, Y)

plt.contourf(X, Y, Z, 20, cmap='OrRd')
plt.colorbar();

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(50, 50)
```

2D and 3D plots for f3:

```
Z = f3(X, Y)

plt.contourf(X, Y, Z, 20, cmap='OrRd')
plt.colorbar();

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
ax.view_init(50, 40)
```

```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
import numpy as np
np.random.seed(42)

x1s = 2 * np.random.rand(100, 1)
x2s = 2 * np.random.rand(100, 1)
ys = 4 + 3 * x1s + 3 * x2s + np.random.rand(100, 1)
```

```
# split the data into training and test sets
# train set
train_x1s = x1s[:80]
train_x2s = x2s[:80]
train_ys = ys[:80]
# test set
test_x1s = x1s[80:]
test_x2s = x2s[80:]
test_ys = ys[80:]
```

```
# number of epochs
epochs = 100
# learning rate
lr = 0.01
```

```
# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_x1s[i] + w2 * train_x2s[i] + b
    
    grad_w1 = (y_pred - train_ys[i]) * train_x1s[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2s[i]
    grad_b1 = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b1
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_x1s[i] + w2 * test_x2s[i] + b - test_ys[i]) ** 2
test_loss /= 20

test_loss
```

```
#using 2 weights to get similar results:
pred_ys = w1 * test_x1s + w2 * test_x2s + b
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(test_x1s, test_x2s, pred_ys, cmap='Greens');
ax.scatter3D(test_x1s, test_x2s, test_ys);
```
